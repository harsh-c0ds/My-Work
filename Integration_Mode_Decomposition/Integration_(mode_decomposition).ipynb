We are considering the following functions:
$$
f_1(x)=e^x\cos(x)
$$
in $I=[0,\pi/2]$,

$$
f_2(x)=e^x
$$
in $I=[-1,3]$,

$$
f_3(x)=
\begin{cases} 
  e^{2x}   &\text{if } x < 0       \\ 
  x-2\cos(x)+4   &\text{if } x\geq0 \\ 
\end{cases} 
$$
in $I=[-1,1]$.

We applied Simpson's 3/8 rule and Gauss-Legendre quadrature. 
Along with Simpson's 3/8 rule we also applied Simpson's 1/3 rule and the trapezoidal rule to deal with the cases MOD(m, 3)=1, and MOD(m, 3)=2, where m is the number of subintervals.

We notice that the error is not converging to zero in the Simpson case. WHY??
Instead, regarding the Gauss-Legendre quadrature, the error approaces zero.

```python
import numpy as np
import matplotlib.pyplot as plt

for item in range(1,4):
    filename = f"doc_{item}.txt"
    data = np.loadtxt(filename)
    x, y = data[:, 0], data[:, 1]

    plt.figure()
    plt.plot(x, y, label="Simpson's 3/8 rule")
    plt.xlabel("log(h)")
    plt.ylabel("Error")
    plt.title(f"Function {item}")
    plt.grid()
    
    filename = f"doc_{item+3}.txt"
    data = np.loadtxt(filename)
    x, y = data[:, 0], data[:, 1]

    plt.plot(x, y, label="Gauss-Legendre quadrature")
    plt.xlabel("log(h)")
    plt.ylabel("Error")
    plt.title(f"Function {item}")
    plt.legend()
    plt.show()

# plot related to Gauss-Legendre only
for item in range(4,7):
    filename = f"doc_{item}.txt"
    data = np.loadtxt(filename)
    x, y = data[:, 0], data[:, 1]

    plt.figure()
    plt.plot(x, y, label="Gauss-Legendre quadrature")
    plt.xlabel("log(h)")
    plt.ylabel("Error")
    plt.title(f"Function {item-3}")
    plt.legend()
    plt.grid()
```

The reason for error not approaching zero in the case of simpson's rule can due to the fact that there are many round off errors arising during the numerical integration. The other reason could be that the Simpson's rule integrates polynomials of degree less than or equal to 3. For more complex functions, residual error could be more. On the orther hand Guass-Legendre Quadrature approximates the function while optimally chosing points and can be usind to integrate polynomials of order 2n - 1. This could be the reason why error approaches zero in the case of Guass-Legendre Poluy nomial and not in the case of Simpson's Rule.

The other methods that could be implemented instead of these to achieve better performance could be Monte-Carlo Integration method. This method approximates the function by choosing several random numbers within the integration limits and perform the integration. Since, this method does not depend upon the degree of polynomial or the weights, we can get the better results by choosing as many random numbers as we can. In this method, the more the random numbers between the given limit the more accurate our answer gets.
